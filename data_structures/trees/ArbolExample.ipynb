```python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.padre = None
        self.hijos = list()

    def asignar_padre(self, padre):
        self.padre = padre
        padre.hijos.append(self)

    def __repr__(self):
        return f"Nodo {self.valor}"

    def profunidad(self):
        activo = self
        profunidad = 0
        while True:
            if not activo.padre:
                return profunidad
            else:
                activo = activo.padre
            profunidad += 1

    def ancestros(self):
        ancestros = []
        activo = self
        while activo.padre:
            activo = activo.padre
            ancestros.append(activo)
        return ancestros

    def ancestros_abajo(self):
        return list(reversed(self.ancestros()))

    def tamaño(self):
        tamaño = 1
        for hijo in self.hijos:
            tamaño += hijo.tamaño()
        return tamaño 
    def altura(self):
        if not self.hijos:
            return 0 
        altura_hijos = [hijos.altura() for hijos in self.hijos]
        return 1 + max(altura_hijos)
        
        
             
```
