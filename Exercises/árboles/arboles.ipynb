```python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.hijos = []
        self.padre = None

    def asignar_padre(self, padre):
        self.padre = padre
        padre.hijos.append(self)

    def distancia_a_raiz(self, cnt=0):
        if not self.padre : return cnt
        else: return self.padre.distancia_a_raiz(cnt+1)

    def get_ancestros(self, ancestros=[]):
        if not self.padre: return ancestros
        else:
            ancestros.append(self.padre)
            return self.padre.get_ancestros(ancestros)

    def tamano(self):
        tam=len(self.hijos)
        if not len(self.hijos): return 0
        else:
            return tam + (h.tamano() for h in self.hijos)
        
    def profundidad(self):
        if not self.hijos: return 0
        else: return 1 + max(h.profundidad() for h in self.hijos)

    def __repr__(self):
        return f'Nodo {self.valor}'

```

```python
A = Nodo("root")
B= Nodo('hijo1')
C = Nodo('hijo2')
B.asignar_padre(A)
C.asignar_padre(A)
D= Nodo("subhijo")
D.asignar_padre(B)

print(A)
print('hijos:',A.hijos)
```
